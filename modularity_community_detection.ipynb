```python
import networkx as nx
import matplotlib.pyplot as plt
from assignment import girvan_newman_communities
```

```python


G = nx.karate_club_graph()
best_partition, best_q, q_history, steps_df = girvan_newman_communities(G)

node_colors = []
for node in G.nodes():
    for i, comm in enumerate(best_partition):
        if node in comm:
            node_colors.append(i)
            break

plt.figure(figsize=(10,7))
nx.draw_spring(G, node_color=node_colors, cmap=plt.cm.tab20, with_labels=True)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(range(len(q_history)), q_history, marker='o')
plt.xlabel('Split Step')
plt.ylabel('Modularity')
plt.grid(True)
plt.show()

print(f"Optimal Modularity: {best_q:.4f}")
print("\nSteps DataFrame:")
print(steps_df.head())
```
